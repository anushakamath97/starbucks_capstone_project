```python
# import required libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
#set seed
np.random.seed(42)
```

```python
# read data from files to dataframe
portfolio_df = pd.read_json("data/portfolio.json", lines=True)
profile_df = pd.read_json("data/profile.json", lines=True)
transcript_df = pd.read_json("data/transcript.json", lines=True)
```

```python
# overview of the dataframes
print("Portfolio Data")
rows, columns = portfolio_df.shape
print("No. of rows:", rows)
print("No. of columns:", columns)

columns = portfolio_df.columns.tolist()
print("Columns:", columns, "\n")

# remove channels as list is unhashable
columns.remove("channels")
columns.remove("id")

for col in columns:
  print("Column:", col)
  print("Unique values:", portfolio_df[col].unique(), "\n")

print("No. of Nan:", "\n", portfolio_df.isna().sum(),"\n\n")
portfolio_df.head()
```

```python
# overview of the dataframes
print("Profile Data")
rows, columns = profile_df.shape
print("No. of rows:", rows)
print("No. of columns:", columns)

columns = profile_df.columns.tolist()
print("Columns:", columns, "\n")

columns.remove("id")

for col in columns:
  print("Column:", col)
  print("Unique values:", profile_df[col].unique(), "\n")
  
print("No. of Nan:", profile_df.isna().sum(),"\n\n")
profile_df.head()
```

```python
# overview of the dataframes
print("Transcript Data")
rows, columns = transcript_df.shape
print("No. of rows:", rows)
print("No. of columns:", columns)

columns = transcript_df.columns.tolist()
print("Columns:", columns, "\n")

columns.remove("person")
columns.remove("value")

for col in columns:
  print("Column:", col)
  print("Unique values:", transcript_df[col].unique(),"\n")

print("No. of Nan:", "\n", transcript_df.isna().sum(),"\n\n")
transcript_df.head()
```

```python
# split channels in portfolio into separate columns
def split_channels(row):
    # web, email, mobile, social
    new_row = [0, 0, 0, 0]
    if "web" in row:
        new_row[0] = 1
    if "email" in row:
        new_row[1] = 1
    if "mobile" in row:
        new_row[2] = 1
    if "social" in row:
        new_row[3] = 1
    return new_row

channels_df = portfolio_df["channels"].apply(split_channels)
channels_df = pd.DataFrame(np.array(channels_df.tolist()), columns=["web", "email", "mobile", "social"])
portfolio_df = pd.concat([portfolio_df.drop("channels", axis=1), channels_df], axis=1)
portfolio_df.head()
```

```python
# convert became_member_on column in profile into timestamp
print("Before conversion:", profile_df["became_member_on"].dtypes)

# first we need to convert int to string and then to datetime
profile_df["became_member_on"] = pd.to_datetime(profile_df["became_member_on"].apply(str), format="%Y%m%d")
print("After conversion:", profile_df["became_member_on"].dtypes)

profile_df.head()
```

```python
# create seaprate columns for each event type in transcript
transcript_df["event"] = transcript_df["event"].apply(lambda x: x.replace(" ", "_"))
transcript_df = pd.concat([pd.get_dummies(transcript_df["event"], prefix="event"),
                           transcript_df.drop("event", axis=1)], axis=1)
transcript_df.head()
```

```python
# find unique keys for value in transcript df
def get_key(value):
    # convert "offer id" in value_type to "offer_id"
    value_type = list(value.keys())[0]
    if value_type == "offer id":
        return "offer_id"
    return value_type

transcript_df["value_type"] = transcript_df["value"].apply(get_key)
print("Unique value types:", transcript_df["value_type"].unique())
```

```python
# find event when value_type is amount
transcript_df[transcript_df["value_type"] == "amount"].head()
```

```python
# remove key from value
def get_value(value_entry):
    return list(value_entry.values())[0]

transcript_df["value"] = transcript_df["value"].apply(get_value)
transcript_df.head()
```

```python
# check for outliers in customer age
sns.boxplot(x=profile_df["age"]);

# rows with age > 100 or age < 15
customer_age = profile_df[(profile_df["age"] > 100) | (profile_df["age"] < 15)]
print("Age > 100 or Age < 15 :", customer_age['age'].unique())

# the number of occurence of both these ages
print("Num. Age = 101:", customer_age[customer_age["age"] == 101].shape[0])
print("Num. Age = 118:", customer_age[customer_age["age"] == 118].shape[0])
```

```python
# we have 2175 customers with age 118, which seems odd.
# Plot the distribution of customer age without age=118
age_df = profile_df[profile_df["age"] != 118]["age"]
age_df.hist();

# We will assign values picked from normal distribution of customer age
age_mean = age_df.mean()
age_stddev = age_df.std()
print("Mean:", age_mean, "Stddev:", age_stddev)
cust_age_dist = np.random.normal(loc=age_mean, scale=age_stddev, size=customer_age[customer_age["age"] == 118].shape[0])

def get_age(age):
    if age == 118:
        return int(np.random.choice(cust_age_dist, replace=False))
    return age
        
profile_df['age'] = profile_df['age'].apply(get_age)
```

```python
profile_df["age"].hist();
print("New mean:", profile_df["age"].mean(), "New stddev:", profile_df["age"].std())
```

```python
# we also have 2175 nan for income and gender

# fill nan in income column with average income
profile_df.loc[profile_df['income'].isnull(), "income"] = profile_df["income"].mean(skipna=True)
```

```python
# fill nan in gender column with random choice from the gender categories
unique_gender = profile_df.gender.unique().tolist()
unique_gender = [g for g in unique_gender if g]

def get_gender(gender):
    if not gender:
        return np.random.choice(unique_gender)
    return gender

profile_df["gender"] = profile_df.gender.apply(get_gender)
```

```python
# save all modified df to csv
portfolio_df.to_csv("data/portfolio.csv")
profile_df.to_csv("data/profile.csv")
transcript_df.to_csv("data/transcript.csv")
```
