```python
# import required libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
# read data from files to dataframe
portfolio_df = pd.read_json("data/portfolio.json", lines=True)
profile_df = pd.read_json("data/profile.json", lines=True)
transcript_df = pd.read_json("data/transcript.json", lines=True)
```

```python
# overview of the dataframes
print("Portfolio Data")
rows, columns = portfolio_df.shape
print("No. of rows:", rows)
print("No. of columns:", columns)

columns = portfolio_df.columns.tolist()
print("Columns:", columns, "\n")

# remove channels as list is unhashable
columns.remove("channels")
columns.remove("id")

for col in columns:
  print("Column:", col)
  print("Unique values:", portfolio_df[col].unique(), "\n")

print("No. of Nan:", "\n", portfolio_df.isna().sum(),"\n\n")
portfolio_df.head()
```

```python
# overview of the dataframes
print("Profile Data")
rows, columns = profile_df.shape
print("No. of rows:", rows)
print("No. of columns:", columns)

columns = profile_df.columns.tolist()
print("Columns:", columns, "\n")

columns.remove("id")

for col in columns:
  print("Column:", col)
  print("Unique values:", profile_df[col].unique(), "\n")
  
print("No. of Nan:", profile_df.isna().sum(),"\n\n")
profile_df.head()
```

```python
# overview of the dataframes
print("Transcript Data")
rows, columns = transcript_df.shape
print("No. of rows:", rows)
print("No. of columns:", columns)

columns = transcript_df.columns.tolist()
print("Columns:", columns, "\n")

columns.remove("person")
columns.remove("value")

for col in columns:
  print("Column:", col)
  print("Unique values:", transcript_df[col].unique(),"\n")

print("No. of Nan:", "\n", transcript_df.isna().sum(),"\n\n")
transcript_df.head()
```

```python
# split channels in portfolio into separate columns
def split_channels(row):
    # web, email, mobile, social
    new_row = [0, 0, 0, 0]
    if "web" in row:
        new_row[0] = 1
    if "email" in row:
        new_row[1] = 1
    if "mobile" in row:
        new_row[2] = 1
    if "social" in row:
        new_row[3] = 1
    return new_row

channels_df = portfolio_df["channels"].apply(split_channels)
channels_df = pd.DataFrame(np.array(channels_df.tolist()), columns=["web", "email", "mobile", "social"])
portfolio_df = pd.concat([portfolio_df.drop("channels", axis=1), channels_df], axis=1)
portfolio_df.head()
```

```python
# convert became_member_on column in profile into timestamp
print("Before conversion:", profile_df["became_member_on"].dtypes)

# first we need to convert int to string and then to datetime
profile_df["became_member_on"] = pd.to_datetime(profile_df["became_member_on"].apply(str), format="%Y%m%d")
print("After conversion:", profile_df["became_member_on"].dtypes)

profile_df.head()
```

```python
# create seaprate columns for each event type in transcript
transcript_df["event"] = transcript_df["event"].apply(lambda x: x.replace(" ", "_"))
transcript_df = pd.concat([pd.get_dummies(transcript_df["event"], prefix="event"),
                           transcript_df.drop("event", axis=1)], axis=1)
transcript_df.head()
```

```python
# find unique keys for value in transcript df
def get_key(value):
    # convert "offer id" in value_type to "offer_id"
    value_type = list(value.keys())[0]
    if value_type == "offer id":
        return "offer_id"
    return value_type

transcript_df["value_type"] = transcript_df["value"].apply(get_key)
print("Unique value types:", transcript_df["value_type"].unique())
```

```python
# find event when value_type is amount
transcript_df[transcript_df["value_type"] == "amount"].head()
```

```python
# remove key from value
def get_value(value_entry):
    return list(value_entry.values())[0]

transcript_df["value"] = transcript_df["value"].apply(get_value)

transcript_df.head()
```

```python
# number of different offers
portfolio_df["offer_type"].value_counts()
```

Questions:

1. People who viewed the offer and completed
2. People who received the offer, did not open but completed
3. What type of offer is most completed
4. Offers sent through which channel gets completed more
5. Which age group of customers tend to complete the offers

```python
# People who viewed the offer and completed

# groupby person and value
offers_df = transcript_df[(transcript_df["event_offer_viewed"] == 1) | (transcript_df["event_offer_completed"] == 1)]
offers_df = offers_df.groupby(by=["person", "value"])
```

```python
offers_completed_user = []
offers_completed_offer = []
num_time_completed = []

# create a dataframe of people who have viewed and completed the the offer
# We don't keep track if they completed the offer when they viewed based on timestamp
for name, group in offers_df:
    if group[group["event_offer_viewed"] == 1].shape[0] > 0 and group[group["event_offer_completed"] == 1].shape[0] > 0:
        offers_completed_user.append(name[0])
        offers_completed_offer.append(name[1])
        num_time_completed.append(min(group[group["event_offer_viewed"] == 1].shape[0], group[group["event_offer_completed"] == 1].shape[0]))
print("Number of people who viewed and completed the offer:", len(offers_completed_user))
```

```python
viewed_completed_df = pd.DataFrame({"person": offers_completed_user, "offer_id": offers_completed_offer, "count": num_time_completed})

# add offer type for each offer id
viewed_completed_df = viewed_completed_df.join(portfolio_df[["id", "offer_type"]].set_index("id"), on="offer_id")

# plot the number of customers who completed the offer after viewing based on offer type
freq_df = viewed_completed_df.groupby("offer_type")["count"].value_counts()
freq_df = freq_df.reset_index(level="offer_type")
freq_df.columns = ["offer_type", "frequency"]
freq_df = freq_df.reset_index()
sns.barplot(x="count", y="frequency", hue="offer_type", data=freq_df);
```

```python
# What type of offer is most completed
offer_completed = transcript_df[transcript_df["event_offer_completed"] == 1]
offer_compl_count = offer_completed["value"].value_counts().reset_index()
offer_compl_count = offer_compl_count.join(portfolio_df[["id", "offer_type"]].set_index("id"), on="index")
sns.barplot(x="offer_type", y="value", data=offer_compl_count);
```

```python
# Which age group of customers tend to complete the offers
offer_completed_age = offer_completed.join(profile_df[["id", "age"]].set_index("id"), on="person")
sns.histplot(data=offer_completed_age, x="age", bins=10);
```
